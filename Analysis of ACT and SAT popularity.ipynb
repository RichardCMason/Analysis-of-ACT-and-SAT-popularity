```python
# result = stats.ttest_ind(experimental, control)
# result.statistic, result.pvalue
```

```python
stats.ttest_ind(act_sat['Participation_act'], act_sat['Participation_sat'])
```

H0: There is no difference between sat and act participation rates

H1: the difference is statistically significant




```python
n=51
act_participation_mean = act_sat['Participation_act'].mean()
act_participation_std_error = np.std(act_sat['Participation_act'], ddof=1)/np.sqrt(n)
```

```python
sat_participation_mean = act_sat['Participation_sat'].mean()
sat_participation_std_error = np.std(act_sat['Participation_sat'], ddof=1)/np.sqrt(n)
```

```python
stats.norm.interval(0.95, sat_participation_mean, sat_participation_std_error)
```

```python
stats.norm.interval(0.95, act_participation_mean, act_participation_std_error)
```
